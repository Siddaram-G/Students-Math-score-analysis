### Gender distribution

```python
plt.figure(figsize = (6,6))
ax = sns.countplot(data = Df, x = "Gender",hue ="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show() # Distriobution of students based on Gender 
```

```python
plt.figure(figsize = (14,7))
sns.boxplot(data = Df,x = "EthnicGroup",y = "MathScore",hue ="Gender")
plt.title("Math score Distribution by gender and Ethnic groups")
plt.show()
```

```python
Df.groupby(["EthnicGroup","Gender"])[["MathScore","ReadingScore","WritingScore"]].mean()
```

```python
sns.set(style="whitegrid")
fig,axes = plt.subplots(1,3,figsize=(12,6))
sns.histplot(data =Df,x="MathScore",kde = True, ax=axes[0], color = "skyblue")
axes[0].set_title("Math score Distribution")

sns.histplot(data =Df,x="ReadingScore",kde = True, ax=axes[1], color = "salmon")
axes[1].set_title("Reading score Distribution")


sns.histplot(data =Df,x="WritingScore",kde = True, ax=axes[2], color = "lightgreen")
axes[2].set_title("Writing score Distribution")

plt.tight_layout()
plt.show()
```

```python
Gb =  Df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
Gb
```

```python
sns.heatmap(Gb, annot = True)
plt.title("Relationship between Parents Education and Students Scores")
plt.show()
```

```python
Gbmst =  Df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
Gbmst
```

```python
sns.heatmap(Gbmst, annot = True)
plt.title("Relationship between Parents Marital Status and Students Scores")
plt.show()
```

```python
# From the above chart we come to khow the the marital status has negligible impact on their scores
```

### Distribution of Ethnic groups

```python
GroupA = Df.loc[(Df['EthnicGroup']== "group A")].count()
GroupB = Df.loc[(Df['EthnicGroup']== "group B")].count()
GroupC = Df.loc[(Df['EthnicGroup']== "group C")].count()
GroupD = Df.loc[(Df['EthnicGroup']== "group D")].count()
GroupE = Df.loc[(Df['EthnicGroup']== "group E")].count()


l = ["group A","group B","group C","group D","group E"]
my_list = [GroupA["EthnicGroup"],GroupB["EthnicGroup"],GroupC["EthnicGroup"],GroupD["EthnicGroup"],GroupE["EthnicGroup"]]
plt.pie(my_list,labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Groups")
plt.show
```
